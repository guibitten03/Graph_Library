```python
from igraph import *
import matplotlib.pyplot as plt
```

```python
g = Graph.Read_Ncol("/content/exemplos_conectividade.ncol", directed=False)
plot(g, target=plt.axes())
```

```python
def scam(g, start, arrival):
  while True: # Péssima prática... Pensar numa solução melhor
    paths = g.get_all_simple_paths(start, arrival)
    if len(paths) == 0: break

    vertices = {}
    for v in g.vs:
      vertices[v.index] = 0

    for l in paths:
      for item in l:
        if item == 0 or item == 10:
          continue
        vertices[item] += 1

    maxV = max(vertices, key=vertices.get)
    g.delete_vertices(maxV)
```

```python
scam(g, 0, 10)
plot(g, target=plt.axes())
```
